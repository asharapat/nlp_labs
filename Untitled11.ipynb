```
START = or_(rule(
    or_(
        rule(gram('ADJF')),
        rule(gram('NOUN')),
        rule(gram('ADVB'))
    ).optional(),
    gram('NOUN'),
    or_(
        rule(gram('ADJF')).repeatable(),
        rule(gram('NOUN'))
    ).optional(),
    gram('NOUN')  
    ),
    rule(gram('NOUN')))

START_S = or_(
    eq('—Ç–∞–∫–æ–π'),
    eq('—Ç–∞–∫–∏–µ'),
)

KAK = eq('–∫–∞–∫')
INCLUDING = or_(
    or_(
        eq('–≤'),
        eq('—Ç–æ–º'),
        eq('—á–∏—Å–ª–µ'),
    ),
    eq('–≤–∫–ª—é—á–∞—é—â–∏–π'),
    or_(
        eq('–≤–∫–ª—é—á–∞—é—â–∏–π'),
        eq('–≤'),
        eq('—Å–µ–±—è'),
    ),
    or_(
        eq('–≤–∫–ª—é—á–∞—é—â–∏–µ'),
        eq('–≤'),
        eq('—Å–µ–±—è'),
    ),
    eq('–≤–∫–ª—é—á–∞—é—â–∏–µ'),
    eq('–æ—Å–æ–±–µ–Ω–Ω–æ'),
    or_(
        eq('–ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
    )
)

MID_S = or_(
    rule(
        or_(
            eq('—Ç–∞–∫–æ–π'),
            eq('—Ç–∞–∫–∏–µ'),
        ),
        eq('–∫–∞–∫')
    )
)
ATAKJE = or_(rule(
    eq(','),
    eq('–∞'),
    eq('—Ç–∞–∫–∂–µ'),
    eq('–≤–µ—Å—å–º–∞')
),
    rule(
    eq('–∏'),
    eq('—Ç–æ–º—É'),
    eq('–ø–æ–¥–æ–±–Ω–æ–≥–æ')
    )
)


MID = or_(
    rule(
        eq('—ç—Ç–æ')
    ),
    rule(
        eq('‚Äî')
    ),
    rule(
        eq('‚Äî'),
        eq('—ç—Ç–æ')
    ),
    rule(
        eq('‚Äî'),
        not_(eq('–∫–º'))
    ),
    rule(
        or_(
            eq('–∏'),
            eq('–∏–ª–∏'),
        ),
        eq('–¥—Ä—É–≥–∏–µ')
    ),
    rule(
        eq('-')
    ),
    rule(
        eq('‚Äì'),
        eq('—ç—Ç–æ'),
        eq('–Ω–µ')
    ),
    rule(
    eq('—è–≤–ª—è–µ—Ç—Å—è')
    ),
    rule(
        eq('-'),
        eq('—ç—Ç–æ')
    ),
    rule(
    or_
        (eq('–Ω–∞–∑—ã–≤–∞—é—Ç'),
        eq('–Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è'),
        eq('–Ω–∞–∑—ã–≤–∞–ª–∏'))
    ),
    rule(
    eq('—Å—á–∏—Ç–∞–µ—Ç—Å—è')
    ),
    rule(eq('—Ç–æ'),
        eq('–µ—Å—Ç—å')),
    rule(
    eq(','),
    eq('–∫–∞–∫')
    ),
        rule(
    eq('–æ–¥–∏–Ω'),
    eq('–∏–∑')
    )
    ,
    rule(
    eq('.'),
    eq('–æ–ø–∏—Å–∞–Ω–∏–µ'),
    eq(':')
    ),
    rule(
    eq('–≤'),
    eq('—Å–ª–æ–≤–∞—Ä–µ'),
    eq('-')
    ),
    rule(
    eq('/'))
    
)

END = or_(
    rule(
        gram('NOUN'),
        gram('NOUN')
    ),
    rule(
        gram('ADJF').repeatable(),
        gram('NOUN')
    ),
    rule(
        gram('ADJF'),
        gram('ADJF').repeatable(),
        gram('NOUN')
    ),
    rule(
        gram('NOUN').repeatable(),
        gram('ADJF'),
        gram('NOUN').repeatable()
    ),
    rule(
        gram('NOUN').repeatable(1,3)
    ), 
    rule(
        gram('ADJF'),
        gram('NOUN')
    ),
    rule(
        gram('ADVB'),
        gram('ADJF'),
        gram('NOUN')
    ),
    rule(    
        gram('ADVB'),
        eq('–∏'),
        gram('PRTF')
    #gram('NOUN').repeatable(1,3)
)
)


Item = fact(
    'Item',
    [attribute('titles').repeatable()]
)
IGNORE = or_(rule(
    '(',
    not_(eq(')')).repeatable(),
    ')'
        ))

ITEM = rule(
    IGNORE.interpretation(
        Item.titles
    ),
    eq(',').optional() 
).repeatable().interpretation(
    Item
)

RULE1 = rule(
    
    gram('ADVB'),
    eq('–∏'),
    gram('PRTF')
    #gram('NOUN').repeatable(1,3)
)

SHORT_WORD = or_(
    rule(
        eq(','),
        eq('-'),
        eq('–∞'),
        eq(','),
        eq('–º'),
        eq('.')
    #, -–∞, –º.
    )
)

ILI = or_(rule(

    eq(','),
    eq('–∏–ª–∏'),
    gram('NOUN'),
    eq(','),
    eq('–∏–ª–∏'),
    gram('NOUN')
),
    rule(
    eq('–∏–ª–∏'),
    gram('NOUN'),
    gram('ADJF')
          ),
    rule(eq('–∏'), 
    gram('NOUN'))
         )
```
